## Topics 

An NMF topic model was run with n=5 componenets. The topics werent obviously or strictly seperated and got increasingly uninterpretable as n got larger. But some initial topic groups could be 

Topic 1: Fuller Body Wines(Top word characteristcs include: Fruits, ripe, acidity, tannins, rich , and wood.) 
Topic 2: Medium Body Wines
Topic 3: Light Body(Mostly seen in whites, with the exception of chardonay which is a dense white)
Topic 4: Still analyzing 
Topic 5: Still analyzing

## Plot 1

Plot # 1 is a side by side comparison of two wines reccomended as similar. The wine on the left was used for comparison against the entirity of my wine/topic vectors. It shares flavor characteristics such as  acitity and fruits and shares texture characteristics such as 'tannins'. These wine vectors were built from a slightly updated corpus. Country of Origin, Region of Origin, and Wine Variety were appended to each document description. 

## Plot 2

Plot # 2 is a distribution of ABV%'s for the 40 most similar wines(still working off the same wine of comparison in plot # 1). Most of the wines have Alcohol By Volume percentages in the 13.5% to 14.5% range. This is indicative of a fuller body seen a lot in red wines. 

## Plot 3

Plot # 3 is a line plot showing the ABV% as we move from most similar to 10th most similar. Two lines are overlaid to indicate industry standard wine body thresholds. From the plot you can see that 8/10 have ABV% in the medium to full body range. The wine of comparison had an ABV% of 14.5%.
